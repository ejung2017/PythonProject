```python
for link in url_list: 
    driver.get(link)
    time.sleep(5)
    try:
        # paragraphs 
        div_element = driver.find_element(By.XPATH, "//div[@class='article-body']")
        article_body = div_element.find_elements(By.TAG_NAME, "p")
        
        content_list = []

        for item in article_body: 
            try: 
                # paragraphs 
                paragraph = item.text
                content_list.append(paragraph)

            except Exception as e:
                print(f"Error extracting data: {e}")
                break 

        content = " ".join(content_list)
        
        # date
        date = driver.find_element(By.CSS_SELECTOR, "div.publish-date").text
        
        #title
        title = driver.find_element(By.CSS_SELECTOR, "h1.title-article").text

        jpt_article.append({
            "link": url_the_japan_times,
            "title": title,
            "date": date,
            "content": content
        })

    except Exception as e:
        print(f"Error extracting data: {e}")
        print(link)
        continue

    # driver.quit()
```
